# Perceptron 

```python
import matplotlib.pyplot as plt
import numpy as np
```

Dataset

```python
x = [[1,-1], [0.3, 5], [-3, 4], [2, 1], [-2,0], [0, 3]]
y = [-1, 1, 1, -1, -1, 1]

x = np.array([[1.0, i[0], i[1]] for i in x])
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.grid()
ax.set_title('DataSet')
for i in range(len(x)):
  marker = 'bo' if y[i] == 1 else '+r'
  plt.plot(x[i][1], x[i][2], marker)
```

The line is given by its affine vector, given as $(a,b,c)$ to indicate the rect $a + bx + cy =0$

```python
w = np.array([8.0, 5.0, 0.0])
eta = 0.1
n = 0


def learn(w):
  run = True
  
  i = 0
  error_found = False
  
  while run:  
    
    while i<len(x) and not error_found: #cycle on all the six values until an error is found
      if y[i]*np.dot(w,x[i]) <= 0.0 :  #if it finds a mislabeled value
        w = w + eta*y[i]*x[i] #correct the value
        error_found = True 
      else:
        i += 1  #otherwise, check the next value
        
    if error_found: #At the end of the cycle, if an error was found start the process again
      i = 0
      error_found = False
    else: #Otherwise, end the run
      run = False       
  return w

w = learn(w)
```
