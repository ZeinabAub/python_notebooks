```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return x**2*np.sin(x)

x = np.linspace(0,10,100)
```

```python
plt.plot(x,f(x))
```

```python
import pandas as pd

data = pd.DataFrame([x,f(x)], index=['x','y'])
data.to_csv('data.csv')
data
```

```python
data1 = pd.read_csv('data.csv')
plt.plot(data1.iloc[0], data1.iloc[1])
plt.show()
```
